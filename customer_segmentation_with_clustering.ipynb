# Feature Engineering

```python
# Create new feature columns to aid concentrated data analysis

# Reference date provided
today = pd.Timestamp('now')

# Removing the dollar sign and commas off assigned values, converts to numeric values & returns NaN if conversion fails
def convert_num(data):
  return pd.to_numeric(data.astype(str).str.replace('[$,]', '', regex=True), errors='coerce')

retail_aggr = (

    # Assigning variables to feature conversions
    retail_data.assign(

        Revenue   = convert_num(retail_data['Total Revenue']),  # Total Revenue is converted to numerical
        Cost      = convert_num(retail_data['Unit Cost']),       # Unit Cost is converted to numerical
        Delivery  = pd.to_datetime(retail_data['Delivery_Date'], format='%d%b%Y'),
        BirthDate = pd.to_datetime(retail_data['Customer_BirthDate'], format='%d%b%Y')   # Delivery_Date and Customer_BirthDate are both converted to real date-time columns
    )

    # Aggregate based on Customer ID
    .groupby('Customer ID').agg(

        Frequency      = ('Order ID', 'count'),
        Recency        = ('Delivery', lambda x: (today - x.max()).days), # Matches latest delivery date with today's date per customer ID
        Avg_Unit_Cost  = ('Cost', 'mean'),
        CLV            = ('Revenue', 'sum'),
        Age            = ('BirthDate', lambda y: (today - y.max()).days // 365)
    )
    .reset_index()
)

retail_aggr.head()
```

### Outliers

```python
# Using IQR to detect outliers in each feature

# Container for outlier summary results
outlier_summary = []

# Loop to include IQR to all features
for col in retail_new.columns:
    sorts = retail_new[col].sort_values()

    # Calculate quantiles
    q1 = sorts.quantile(0.25)
    q3 = sorts.quantile(0.75)
    iqr = q3 - q1

    # Calculate the upper & lower limits
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr

    # Calculate total outliers and percentage
    outlier = (sorts < lower) | (sorts > upper)
    outlier_total = outlier.sum()                   # Calculates total number of outliers
    outlier_pct = outlier_total / len(sorts) * 100  # Calculates percentage of outliers in total

    outlier_summary.append({'Feature': col,
                    'Outliers': outlier_total,
                    'Percentage of total': f'{round(outlier_pct, 2)}%'})

# Summary table of features and their outliers
outlier_table = pd.DataFrame(outlier_summary).set_index('Feature')
outlier_table
```

The data table above confirms the observations from the boxplots. All features except Age contain statistical outliers.

# Feature Scaling

```python

# Using ColumnTransformer to ensure only numeric columns are scaled and guard against future changes

numerical_features = ['Frequency', 'Recency', 'CLV', 'Avg_Unit_Cost', 'Age']

# Pipeline for preprocessing
preprocessor = ColumnTransformer(
transformers=[
('num', StandardScaler(), numerical_features)])

# Fit and transform the data
retail_preprocessed = preprocessor.fit_transform(retail_new)

# Convert the preprocessed data back to a DataFrame
retail_preprocessed_df = pd.DataFrame(retail_preprocessed, columns=numerical_features)

retail_preprocessed_df.head()
```

# K-Means pipeline

```python
'''
Using k-means to set clusters (which would be determined later)
Using Pipeline to chain multiple estimators into one.
The Pipeline used ensures that both the scaling preprocessing and the clustering are remembered everytime its called upon
'''

# Segmentation with K-Means

kmeans = KMeans(n_clusters= 3, random_state=42)

# Creating the pipeline that preprocesses the data and applies K-Means
pipeline = Pipeline(steps=[('preprocessor', preprocessor), ('clusterer', kmeans)])

# Fit the pipeline to the data
pipeline.fit(retail_new)

# Get the cluster assignments
cluster_labels = pipeline.predict(retail_new)
```

# Clustering Methods

## Elbow Method

```python
# Using the elbow method to visualise the optimum number of clusters

# Elbow method with pipeline
wcss=[]
for i in range(1, 11):
    kmeans_elbow = KMeans(n_clusters=i, init = 'k-means++', random_state = 0, n_init = 10)
    kmeans_elbow.fit(retail_preprocessed)
    wcss.append(kmeans_elbow.inertia_)

plt.plot(range(1,11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

The Elbow method was used to visualise the dataset clustering and derive the optimum number of clusters.

Based on the Elbow plot, there is no clear bend/elbow in the diagram.

This indicates that further methods, including the silhouette score method & heirachical clustering method, should be used in conjuction to derive a conclusion on the optimum number of clusters

Between 4-5 clusters can be estimated as the optimum clusters from the plot

## Silhouette Score Method

```python
range_n_clusters = [2, 3, 4, 5, 6, 7]

for n_clusters in range_n_clusters:

    # Creating a subplot with 1 row and 2 columns.

    fig, (ax1, ax2) = plt.subplots(1, 2)
    fig.set_size_inches(18, 7)

    # The 1st subplot is the silhouette plot.

    ax1.set_xlim([-0.1, 1])
    '''
    The (n_clusters+1)*10 is for inserting blank space between silhouette
    plots of individual clusters, to demarcate them clearly.
    '''
    ax1.set_ylim([0, len(retail_preprocessed) + (n_clusters + 1) * 10])

    '''
    Initialize the clusterer with n_clusters value and a random generator
    seed of 10 for reproducibility.
    '''
    clusterer = KMeans(n_clusters=n_clusters, random_state=10, n_init = 10)
    cluster_labels = clusterer.fit_predict(retail_preprocessed)

    # The silhouette_score gives the average value for all the samples.
    silhouette_avg = silhouette_score(retail_preprocessed, cluster_labels)
    print("For n_clusters =", n_clusters,
          "The average silhouette_score is :", silhouette_avg)

    # Computing the silhouette scores for each sample
    sample_silhouette_values = silhouette_samples(retail_preprocessed, cluster_labels)

    y_lower = 10
    for i in range(n_clusters):
        # Aggregate the silhouette scores for samples belonging to cluster i, and sort them.
        ith_cluster_silhouette_values = \
            sample_silhouette_values[cluster_labels == i]

        ith_cluster_silhouette_values.sort()

        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i

        color = cm.nipy_spectral(float(i) / n_clusters)
        ax1.fill_betweenx(np.arange(y_lower, y_upper),
                          0, ith_cluster_silhouette_values,
                          facecolor=color, edgecolor=color, alpha=0.7)

        # Labeling the silhouette plots with their cluster numbers at the middle.
        ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))

        # Computing the new y_lower for next plot.
        y_lower = y_upper + 10  # 10 for the 0 samples.

    ax1.set_title("The silhouette plot for the various clusters.")
    ax1.set_xlabel("The silhouette coefficient values")
    ax1.set_ylabel("Cluster label")

    # The vertical line for average silhouette score of all the values.
    ax1.axvline(x=silhouette_avg, color="red", linestyle="--")

    ax1.set_yticks([])  # Clearing the yaxis labels / ticks.
    ax1.set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])

    # 2nd Plot showing the actual clusters formed.
    colors = cm.nipy_spectral(cluster_labels.astype(float) / n_clusters)
    ax2.scatter(retail_preprocessed[:, 0], retail_preprocessed[:, 1], marker='.', s=30, lw=0, alpha=0.7,
                c=colors, edgecolor='k')

    # Labeling the clusters.
    centers = clusterer.cluster_centers_

    # To draw white circles at cluster centers.
    ax2.scatter(centers[:, 0], centers[:, 1], marker='o',
                c="white", alpha=1, s=200, edgecolor='k')

    for i, c in enumerate(centers):
        ax2.scatter(c[0], c[1], marker='$%d$' % i, alpha=1,
                    s=50, edgecolor='k')

    ax2.set_title("The visualization of the clustered data.")
    ax2.set_xlabel("Feature space for the 1st feature")
    ax2.set_ylabel("Feature space for the 2nd feature")

    plt.suptitle(("Silhouette analysis for KMeans clustering on sample data "
                  "with n_clusters = %d" % n_clusters),
                 fontsize=14, fontweight='bold')

plt.show()
```

Based on the Silhouette scores and plot generated above, n_clusters = 5 appears to be the ideal number of clusters, as it holds the highest silhouette score (0.26711), there is a stable distribution across all clusters, despite the presence of outliers. It also has more compact clusters with less variability amonst its points comapred to n_clusters = 2, which has a closer silhoutte score.

It is also observed that in all clusters, we have some silhoutte scores falling below zero, n_clusters = 7, has the most below zero (5). This indicates the presecnce of outliers in the data set, as indicated in earlier analysis methods.

Despite having a clearer understanding of the optimum number of clusters to be used using the silhouette score method than the elbow method, further methods can be used to derive a clearer conclusion.

```python
# Merge PCA data with cluster column

y = retail_old['Cluster']
retail_pca['Cluster'] = y

retail_pca.columns = ['PC1', 'PC2', 'Cluster']
retail_pca
```

```python
# Visualize PCA plot

plt.figure(figsize=(8, 6))

# Generate color palette
palette = ['#e41a1c', '#377eb8', '#4daf4a', '#984ea3', '#ff7f00']
colors = [palette[i] for i in retail_pca['Cluster']]

# PCA plot
sns.scatterplot(data=retail_pca,
                x='PC1', y='PC2',
                hue='Cluster',
                palette=colors, s=60, alpha=0.7)

plt.title('PCA Results of Customer Segments')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend(title='Cluster')
plt.tight_layout(); plt.show()
```

The PCA plot above indicates the 5 cluster points using PC1 and PC2.

Despite some overlaps in the clustering, we observe the significant seperation among the clusters confirming our earlier analysis.

Note that there is an extreme case at the top left of the chart. This could be an extreme/one-off case, and requires further investigation and analysis.

# t-SNE (Non-linear dimensionality reduction)

```python
# Visualize t-SNE plot

plt.figure(figsize=(12, 6))

for idx, p in enumerate([15, 25, 30]):

    tsne = TSNE(n_components=2, perplexity=p, random_state=0)
    tsne_results = tsne.fit_transform(retail_preprocessed)

    # Merge t-SNE data with cluster column
    tsne_results = pd.DataFrame(tsne_results, columns=['tsne1', 'tsne2'])
    tsne_results['Cluster'] = retail_old['Cluster'].values


    # Plot t-SNE plots
    plt.subplot(2, 3, idx+1)
    sns.scatterplot(data=tsne_results, x='tsne1', y='tsne2', hue='Cluster',
                    palette='Set2', s=30, alpha=0.8, legend=False)
    plt.title(f'Perplexity = {p}')
    plt.xlabel(''); plt.ylabel('')

plt.tight_layout()
plt.show()
```

We observed different perplexities in order to view different lenses of the clustering plot.

From the plot below (Figure 9), perplexity = 15 reflects the tightest, most compact and most differentiated clusters out of the 3 perplexities.
